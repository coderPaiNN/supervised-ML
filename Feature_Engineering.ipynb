## one hot encoding for the top 10 most frequent categories as 1 and other categories as 0

```python
import pandas as pd
import datetime
```

```python
# First we will create variable with dates and from that extract the week days
# create a list with dates 20 days difference from today 
# then transform it into dataframe
```

```python
df_base = datetime.datetime.today()
```

```python
df_base
```

```python
df_list = [df_base - datetime.timedelta(days=x) for x in range(20)]
```

```python
df_list
```

```python
df = pd.DataFrame(df_list,columns=['day'])
```

```python
df
```

```python
df['week_day_name'] = df['day'].dt.day_name()
```

```python
df
```

```python
week_day_map = {'Monday':1,
                'Tuesday':2,
                'Wednesday':3,
                'Thursday':4,
                'Friday':5,
                'Saturday':6,
                'Sunday':7}
```

```python
df['day_ordinal'] = df.week_day_name.map(week_day_map)
```
